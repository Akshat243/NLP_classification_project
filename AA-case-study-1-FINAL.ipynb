## 4.1 Useful representations for NLP

AI models especially DL using neural networks require numerical representations to learn representations.  Using plain text input is limited to text processors and parsers that will be limited by "rules-based" approaches.

Also, here is where deciding what to process between the blog titles or summaries and what a useful *chunk* of information will be to consider as model input.

Use your skills to further transform your text inputs into representations that can fit a NLP DL model and pipeline.

- HINT:  Stopwords, Pre-trained models, Vocabulary map, Tokenization, Embeddings

```
### 4.1.1 - Insert working cells here
from sklearn import model_selection, preprocessing, metrics
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn import decomposition, ensemble

import xgboost, textblob, string
from keras.preprocessing import text, sequence
from keras import layers, models, optimizers
# split the dataset into training and validation datasets 
train_x, valid_x, train_y, valid_y = model_selection.train_test_split(newdata['summary'].values.astype('U'), newdata['category'])
valid2_x, test_x, valid2_y, test_y = model_selection.train_test_split(valid_x, valid_y)
# label encode the target variable 
encoder = preprocessing.LabelEncoder()
train_y = encoder.fit_transform(train_y)
valid_y = encoder.fit_transform(valid_y)
test_y = encoder.fit_transform(test_y)
```

```
# word level tf-idf
tfidf_vect = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}', max_features=5000)
tfidf_vect.fit(newdata['summary'].values.astype('U'))
xtrain_tfidf =  tfidf_vect.transform(train_x)
xvalid_tfidf =  tfidf_vect.transform(valid_x)
xtest_tfidf = tfidf_vect.transform(test_x)

# ngram level tf-idf 
tfidf_vect_ngram = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}', ngram_range=(2,3), max_features=5000)
tfidf_vect_ngram.fit(newdata['summary'].values.astype('U'))
xtrain_tfidf_ngram =  tfidf_vect_ngram.transform(train_x)
xvalid_tfidf_ngram =  tfidf_vect_ngram.transform(valid_x)
xtest_tfidf_ngram = tfidf_vect_ngram.transform(test_x)
```

<a name="4-2"></a>
## 4.2 On the shoulder of giants, Re-usable representations

The latest NLP innovation trends are related to transformer architectures.  Some of these approaches power 1 out 10 of our current Google searches.

Although are not required to use such approach and can propose other NLP DL model architectures inspired by image process (i.e., CNNs) or sequence models for machine translation, etc. (i.e., RNNs, etc.) each have their own tradeoffs.

Try to implement one that make sense given the time you have.

Trying more than one will definitely highlight your problem solving ability.

- HINT: It may be possible that you have to revisit your dependencies and libraries as you may have to acquire more remote files (e.g., stop words, etc.) for downstream tasks.

```
### 4.2.1 - Insert working cells here
def create_model_architecture(input_size):
    # create input layer 
    input_layer = layers.Input((5000, ))
    
    # create hidden layer
    hidden_layer1 = layers.Dense(100, activation="relu")(input_layer)
    hidden_layer2 = layers.Dense(200, activation="relu")(hidden_layer1)
    hidden_layer3 = layers.Dense(100, activation="relu")(hidden_layer2)

    # create output layer
    output_layer = layers.Dense(1, activation="sigmoid")(hidden_layer3)

    classifier = models.Model(inputs = input_layer, outputs = output_layer)
    classifier.compile(optimizer=optimizers.Adam(), loss='binary_crossentropy')
    return classifier 
```

<a name="4-3"></a>
## 4.3 Create a model for NLP classifier

Based on thinking through sections 4.1, and 4.2 now let's get more tactical.

Using your skills, create a model that predict target classes of 0 or 1.

```
### 4.3.1 - Insert working cells here
def train_model(classifier, feature_vector_train, label, feature_vector_valid, batchsize, epoch):
    # fit the training dataset on the classifier
    history=classifier.fit(feature_vector_train,label,epochs=epoch,batch_size=batchsize,validation_data=(feature_vector_valid, valid_y))
    
    # predict the labels on validation dataset
    predictions = classifier.predict(feature_vector_valid)    
    predictions = predictions.argmax(axis=-1)
    
    return history,metrics.accuracy_score(predictions, valid_y) 
```

<a name="4-4"></a>
## 4.4 Train Model

Once you have constructure your model.  Now it is time to train it.

Use your skills to train the model.  Pick a reasonabile number of epochs and plan to show that training can be outputted in a cell while processing.

Take into consideration of configuring your model for things necessary in downstream tasks.

For example:
- Capturing key metrics during training that can be use for visualization to expain model performance
- Early stopping; in case the model is not performing well
- Optimizing your notebook for model training performance


```
### 4.4.1 - Insert working cells here
epoch=15
batchsize=10
classifier = create_model_architecture(xtrain_tfidf_ngram.shape[0])
history,accuracy = train_model(classifier, xtrain_tfidf_ngram, train_y, xvalid_tfidf_ngram,epoch=epoch,batchsize=batchsize)
print ("Validation Accuracy",  accuracy)
```
